# LEARNING

```python
X=project.iloc[:,[4,6,18]].values
y=project.iloc[:,2].values
```

```python
model3=sm.OLS(y,X).fit()
model3.summary()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

# CLASSIFICATION

## LOGISTIC REGRESSION

```python
from sklearn.linear_model.logistic import LogisticRegression
lgsr=LogisticRegression(random_state=0)
lgsr.fit(X_train,y_train)
```

```python
y_prediction=lgsr.predict(X_test)
y_prediction
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix=confusion_matrix(y_test,y_prediction)
conf_matrix
```

```python
from sklearn.metrics import accuracy_score
accuracy_lgsr=accuracy_score(y_prediction,y_test)
print("accuracy using LOGISTIC REGRESSION is:",accuracy_lgsr)
```

## DECISION TREES

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
y_pred=dtc.predict(X_test)
y_pred
```

```python
conf_matrix=confusion_matrix(y_pred,y_test)
conf_matrix
```

```python
from sklearn.metrics import accuracy_score
accuracy_dtc=accuracy_score(y_pred,y_test)
print("accuracy using DECISION TREE CLASSIFIER is:",accuracy_dtc)
```

## KNEAREST NEIGHBORS

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier(n_neighbors=5)
knc.fit(X_train,y_train)
```

```python
y_predict=knc.predict(X_test)
y_predict
```

```python
conf_matrix=confusion_matrix(y_predict,y_test)
conf_matrix
```

```python
accuracy_knc=accuracy_score(y_predict,y_test)
print("accuracy using KNN CLASSIFIER is:",accuracy_knc)
```
